```python
# Converting the Date column to datetime
df['date'] = pd.to_datetime(df['Date'])

# Rounding of the gold price
df['Gold_Price_USD_YFinance'] = df['Gold_Price_USD_YFinance'].round(2)

# definition of Date as a temporal index
df.set_index('Date', inplace=True)
```

```python
# Renaming of columns
df = df.rename(columns={'Date': 'date', 'Gold_Price_USD_YFinance': 'gold_price_usd'})
```

```python
# Create of columns(Feature engineering)
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['year_month'] = df['date'].dt.to_period('M').astype(str)
df['decade'] = (df['year'] // 10) * 10
df['period_type'] = df['decade'].astype(str) + 's'
```

```python
df.head()
```

```python
# Reasoned removal of the variables not retained
df = df.drop(columns=['month'])
```

```python
# convert to excel
df.to_excel("gold_price_2000_2026_feb.xlsx",index=False,sheet_name="gold_price_2000_2026_feb")
```
